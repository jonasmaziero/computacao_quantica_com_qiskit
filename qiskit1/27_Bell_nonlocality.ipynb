# Teorema de Bell

## Desigualdade CHSH
Em 1964 John Bell tentou mostrar que EPR estavam certos, mas com uma abordagem que possibilitaría fazer testes experimentais. Bell supôs que existiriam variáveis ocultas, tais que se conhecêssemos seus valores, ou a distribuição de probabilidades desses valores, poderíamos prever com certeza os resultados de qualquer medida em MQ. Essa suposição pode ser chamada de _determinismo_. Bell também assumiu _localidade_, ou seja, que o resultado da medida em um sistema não pode influenciar os resultados de medidas no outro sistema. Agora, seguindo Clauser, Horne, Shimoni e Holt, consideramos o seguinte observável
\begin{align}
O & = A_{1}B_{1} + A_{1}B_{2} + A_{2}B_{1} - A_{2}B_{2} \\
  & = A_{1}(B_{1} + B_{2}) + A_{2}(B_{1} - B_{2}),
\end{align}
com $A_{j}$ e $B_{k}$ sendo observáveis dicotômicos, todos tendo espectro $\pm 1$, dos subsistemas $A$ e $B$, respectivamente. Segundo as suposições acima, o valor médio de $O$ ficaria
\begin{equation}
\langle O\rangle_{dl} = \int d\lambda\rho(\lambda) \left\{a_{1}(\lambda)[b_{1}(\lambda) + b_{2}(\lambda)] + a_{2}(\lambda)[b_{1}(\lambda) - b_{2}(\lambda)]\right\},
\end{equation}
em que denotamos por $a_{j},b_{k}$ os possíveis valores de $A_{j},B_{k}$ e $\rho(\lambda)$ é a distribuição de probabilidades para as variáveis ocultas. Note que a suposição de determinismo entra na dependência dos valores dos observáveis com $\lambda$ e a localidade entra na não dependência desses valores com aqueles do outro sistema. O contrário, i.e., a não localidade, seria e.g. $a_{j}=a_{j}(\lambda,b_{k})$. Considerando todas as combidacões possíveveis de valores de $A_{j}$ e $B_{k}$ vemos que o espectro de $O$ é $\pm2$. Disso segue que o valor médio desse observável deve estar limitado por
\begin{equation}
-2 \le \langle O\rangle_{dl} \le 2.
\end{equation}

Agora consideremos um situação idêntica mas descrita pelas regras da MQ. Usaremos como estado preparado $|\Psi_{-}\rangle_{AB}$. Quanto aos observáveis, uma escolha apropriada é
\begin{align}
& A_{1}=\sigma_{z}^{A}, \\
& A_{2}=\sigma_{x}^{A}, \\
& B_{1}=(\sigma_{z}^{B}+\sigma_{x}^{B})/\sqrt{2}, \\ 
& B_{2}=(\sigma_{z}^{B}-\sigma_{x}^{B})/\sqrt{2}.
\end{align}
Pode-se verificar que nesse caso
\begin{equation}
O=\sqrt{2}(\sigma_{z}\otimes\sigma_{z}+\sigma_{x}\otimes\sigma_{x}).
\end{equation}
Ademais, usando $\sigma_{z}^{S}|z_{+}^{S}\rangle=|z_{+}^{S}\rangle$ e $\sigma_{z}^{S}|z_{-}^{S}\rangle=-|z_{-}^{S}\rangle$, teremos que 
\begin{align}
\sigma_{z}^{A}\otimes\sigma_{z}^{B}|\Psi_{-}\rangle_{AB} &= \frac{1}{\sqrt{2}}\left(\sigma_{z}^{A}|z_{+}\rangle_{A}\otimes\sigma_{z}^{B}|z_{-}\rangle_{B} - \sigma_{z}^{A}|z_{-}\rangle_{A}\otimes\sigma_{z}^{B}|z_{+}\rangle_{B}\right) \\  & = \frac{1}{\sqrt{2}}\left(|z_{+}\rangle_{A}\otimes(-|z_{-}\rangle_{B}) - (-|z_{-}\rangle_{A})\otimes|z_{+}\rangle_{B}\right) \\
& = -|\Psi_{-}\rangle_{AB}.
\end{align}
Pode-se verificar também que $\sigma_{x}\otimes\sigma_{x}|\Psi_{-}\rangle_{AB}=-|\Psi_{-}\rangle_{AB}$. Assim,
\begin{align}
\langle O\rangle_{mq} & = _{AB}\langle\Psi_{-}|O|\Psi_{-}\rangle_{AB} = _{AB}\langle\Psi_{-}|\sqrt{2}(-2|\Psi_{-}\rangle_{AB}) \\
                      & = -2\sqrt{2} < -2.
\end{align}
Ou seja, o resultado da MQ está fora do limite imposto pela teoria de variáveis ocultas locais. Desde 1982, vários experimentos foram feitos fornecendo resultados compatíveis com a MQ. Isso nos mostra que uma ou as duas suposições estão erradas, i.e., a Natureza é não determinística e/ou não local. Bell preferia esta última opção, pois existe uma versão conhecida da MQ, a de de Broglie-Bohm, que é determinística e não local. 

## Verificação no IBMQ
A seguir, faremos experimentos no IBMQ para testar a desigualdade de CHSH-Bell.

Vimos que para medir um observável qualquer, $\hat{n}\cdot\vec{\sigma} = \sin\theta\cos\phi\sigma_{x} + \sin\theta\sin\phi\sigma_{y} + \cos\theta\sigma_{z} = \begin{bmatrix}
\cos\theta & e^{-i\phi}\sin\theta \\ e^{i\phi}\sin\theta & -\cos\theta \\ \end{bmatrix}$, de um qubit precisamos aplicar $(U(\theta,\phi,\pi-\phi))^{\dagger} = U(\theta,\phi,\pi-\phi)=\begin{bmatrix} \cos(\theta/2) & e^{-i\phi}\sin(\theta/2) \\ e^{i\phi}\sin(\theta/2) & -\cos(\theta/2) \end{bmatrix}$ no estado preparado. Para medir os observáveis acima, teremos as seguintes unitárias
\begin{align}
& A_{1} = \sigma_{z} \ \therefore\ U(\theta=0, \phi = 0, \lambda = \pi) = Z, \\
& A_{2} = \sigma_{x} \ \therefore\ U(\theta=\pi/2, \phi = 0, \lambda = \pi) = H, \\
& B_{1} = 2^{-1/2}(\sigma_{z}+\sigma_{x}) \ \therefore\ U(\theta=\pi/4, \phi = 0, \lambda = \pi), \\
& B_{2} = 2^{-1/2}(\sigma_{z}-\sigma_{x}) \ \therefore\ U(\theta=-\pi/4, \phi = 0, \lambda = \pi).
\end{align}

Observamos que
\begin{align}
\langle\Psi_{-}|\sigma_{z}\otimes\sigma_{z}|\Psi_{-}\rangle & = -\langle\Psi_{-}|\Psi_{-}\rangle = -1, \\
\langle\Psi_{-}|\sigma_{x}\otimes\sigma_{x}|\Psi_{-}\rangle & = -\langle\Psi_{-}|\Psi_{-}\rangle = -1, \\
\langle\Psi_{-}|\sigma_{z}\otimes\sigma_{x}|\Psi_{-}\rangle & = \langle\Psi_{-}|\Phi_{+}\rangle = 0, \\
\langle\Psi_{-}|\sigma_{x}\otimes\sigma_{z}|\Psi_{-}\rangle & = -\langle\Psi_{-}|\Phi_{+}\rangle = 0.
\end{align}
Por conseguinte
\begin{align}
& \langle A_{1}\otimes B_{1}\rangle_{\Psi_{-}} = -1/\sqrt{2}, \\
& \langle A_{1}\otimes B_{2}\rangle_{\Psi_{-}} = -1/\sqrt{2}, \\
& \langle A_{2}\otimes B_{1}\rangle_{\Psi_{-}} = -1/\sqrt{2}, \\
& \langle A_{2}\otimes B_{2}\rangle_{\Psi_{-}} = 1/\sqrt{2}.
\end{align}

```python
from qiskit import QuantumCircuit
def qc_state_prep():
    qc = QuantumCircuit(2, name = 'state_prep'); 
    qc.h(0)
    qc.cx(0, 1)
    qc.z(0)
    qc.x(1)
    return qc
qc_state_prep_ = qc_state_prep(); qc_state_prep_.draw(output = 'mpl')
```

## Efeito do emaranhamento
Consideremos os mesmos observáveis que anteriormente mas um estado com quantidade ajustável de emaranhamento:
$$|\Psi\rangle = \sqrt{p}|01\rangle - \sqrt{1-p}|10\rangle,$$
com $p\in[0,1]$. Nesse caso temos que
\begin{align}
\langle Z\otimes Z\rangle_{\Psi} & = \langle\Psi|Z\otimes Z\big(\sqrt{p}|01\rangle-\sqrt{1-p}|10\rangle\big) = \big(\sqrt{p}\langle 01|-\sqrt{1-p}\langle 10|\big)\big(-\sqrt{p}|01\rangle+\sqrt{1-p}|10\rangle\big) \\
& = -p -(1-p) = -1, \\
\langle Z\otimes X\rangle_{\Psi} & = \langle\Psi|Z\otimes X\big(\sqrt{p}|01\rangle-\sqrt{1-p}|10\rangle\big) = \big(\sqrt{p}\langle 01|-\sqrt{1-p}\langle 10|\big)\big(\sqrt{p}|00\rangle+\sqrt{1-p}|11\rangle\big) \\
& = 0 = \langle X\otimes Z\rangle_{\Psi}, \\
\langle X\otimes X\rangle_{\Psi} & = \langle\Psi|X\otimes X\big(\sqrt{p}|01\rangle-\sqrt{1-p}|10\rangle\big) = \big(\sqrt{p}\langle 01|-\sqrt{1-p}\langle 10|\big)\big(\sqrt{p}|10\rangle-\sqrt{1-p}|01\rangle\big) \\
& = -2\sqrt{p(1-p)}.
\end{align}

Por conseguinte
\begin{align}
& \langle A_{1}\otimes B_{1}\rangle_{\Psi} = -1/\sqrt{2}, \\
& \langle A_{1}\otimes B_{2}\rangle_{\Psi} = -1/\sqrt{2}, \\
& \langle A_{2}\otimes B_{1}\rangle_{\Psi} = -2\sqrt{p(1-p)}/\sqrt{2}, \\
& \langle A_{2}\otimes B_{2}\rangle_{\Psi} = 2\sqrt{p(1-p)}/\sqrt{2}.
\end{align}
Assim
\begin{align}
\langle O\rangle_{\Psi} & = \langle A_{1}\otimes B_{1}\rangle_{\Psi} + \langle A_{1}\otimes B_{2}\rangle_{\Psi} + \langle A_{2}\otimes B_{1}\rangle_{\Psi} - \langle A_{2}\otimes B_{2}\rangle_{\Psi} \\
& = -\frac{1}{\sqrt{2}} -\frac{1}{\sqrt{2}} - \frac{2\sqrt{p(1-p)}}{\sqrt{2}} - \frac{2\sqrt{p(1-p)}}{\sqrt{2}} \\
& = -\sqrt{2}\big(1+2\sqrt{p(1-p)}\big).
\end{align}
Se $\sqrt{p}=\cos(\theta/2)$ e $\sqrt{1-p}=\sin(\theta/2)$ teremos
$$\langle O\rangle_{\Psi} = -\sqrt{2}\big(1+2\sin(\theta/2\cos(\theta/2\big) = -\sqrt{2}\big(1+\sin\theta\big).$$

OBS. Emaranhamento está de certo modo relacionado com não localidade de Bell. Para estados bipartidos puros, estas duas quanticidades são equivalentes. Para estados mistos, emaranhamento é um condição necessária mas não suficiente para não localidade de Bell.

```python
p = np.arange(0, math.pi/2+0.1, 0.1); Oavg = -math.sqrt(2)*(1+np.sin(p))
d = p.shape[0]; y = -2*np.ones(d); Oavg_sim = np.zeros(d); Oavg_exp = np.zeros(d)
for j in range(0, d):
    # A1B1
    qc = QuantumCircuit(2, 2)
    qc_state_prep_ = qc_state_prep(p[j]); qc.append(qc_state_prep_, [0, 1])
    qc_A1B1_ = qc_A1B1(); qc.append(qc_A1B1_, [0, 1]); qc.measure([0, 1], [0, 1])
    jobS = execute(qc, backend = simulator, shots = nshots)
    resultS = jobS.result().get_counts(); A1B1_avg = AB_avg(resultS, nshots)
    jobE = execute(qc, backend = device, shots = nshots); job_monitor(jobE)
    A1B1_avgE = AB_avg(jobE.result().get_counts(), nshots)
    # A1B2
    qc = QuantumCircuit(2, 2)
    qc_state_prep_ = qc_state_prep(p[j]); qc.append(qc_state_prep_, [0, 1])
    qc_A1B2_ = qc_A1B2(); qc.append(qc_A1B2_, [0, 1]); qc.measure([0, 1], [0, 1])
    jobS = execute(qc, backend = simulator, shots = nshots)
    resultS = jobS.result().get_counts(); A1B2_avg = AB_avg(resultS, nshots)
    jobE = execute(qc, backend = device, shots = nshots); job_monitor(jobE)
    A1B2_avgE = AB_avg(jobE.result().get_counts(), nshots)
    # A2B1
    qc = QuantumCircuit(2, 2)
    qc_state_prep_ = qc_state_prep(p[j]); qc.append(qc_state_prep_, [0, 1])
    qc_A2B1_ = qc_A2B1(); qc.append(qc_A2B1_, [0, 1]); qc.measure([0, 1], [0, 1])
    jobS = execute(qc, backend = simulator, shots = nshots)
    resultS = jobS.result().get_counts(); A2B1_avg = AB_avg(resultS, nshots)
    jobE = execute(qc, backend = device, shots = nshots); job_monitor(jobE)
    A2B1_avgE = AB_avg(jobE.result().get_counts(), nshots)
    # A2B2
    qc = QuantumCircuit(2, 2)
    qc_state_prep_ = qc_state_prep(p[j]); qc.append(qc_state_prep_, [0, 1])
    qc_A2B2_ = qc_A2B2(); qc.append(qc_A2B2_, [0, 1]); qc.measure([0, 1], [0, 1])
    jobS = execute(qc, backend = simulator, shots = nshots)
    resultS = jobS.result().get_counts(); A2B2_avg = AB_avg(resultS, nshots)
    jobE = execute(qc, backend = device, shots = nshots); job_monitor(jobE)
    A2B2_avgE = AB_avg(jobE.result().get_counts(), nshots)
    # O
    Oavg_sim[j] = A1B1_avg + A1B2_avg + A2B1_avg - A2B2_avg
    Oavg_exp[j] = A1B1_avgE + A1B2_avgE + A2B1_avgE - A2B2_avgE
```

## Efeito de observáveis locais
Vamos continuar usando 
$$A_{1} = \sigma_{z} \text{ e } A_{2} = \sigma_{x}.$$
No entanto, vamos escolher os observáveis de Bob tais que 
$$B_{1} = \cos\theta\sigma_{z} + \sin\theta\sigma_{x} \text{ e } B_{2} = 2^{-1/2}(\sigma_{z}-\sigma_{x}),$$
com $\theta\in[-\pi/4,\pi/4]$. Vamos fixar o estado em 
$$|\Psi_{-}\rangle.$$
Lembrando que $\langle Z\otimes Z\rangle_{\Psi_{-}} = \langle X\otimes X\rangle_{\Psi_{-}} = -1$ e $\langle Z\otimes X\rangle_{\Psi_{-}} = \langle X\otimes Z\rangle_{\Psi_{-}} = 0$, obtemos
\begin{align}
\langle A_{1}\otimes B_{1}\rangle_{\Psi_{-}} & = -\cos\theta, \\
\langle A_{1}\otimes B_{2}\rangle_{\Psi_{-}} & = -1/\sqrt{2}, \\
\langle A_{2}\otimes B_{1}\rangle_{\Psi_{-}} & = -\sin\theta, \\
\langle A_{2}\otimes B_{2}\rangle_{\Psi_{-}} & = 1/\sqrt{2}.
\end{align}
Com isso, vem que
\begin{align}
\langle O\rangle_{\Psi_{-}} & = \langle A_{1}\otimes B_{1}\rangle_{\Psi_{-}} + \langle A_{1}\otimes B_{2}\rangle_{\Psi_{-}} + \langle A_{2}\otimes B_{1}\rangle_{\Psi_{-}} - \langle A_{2}\otimes B_{2}\rangle_{\Psi_{-}} \\
& = -2/\sqrt{2}-\cos\theta-\sin\theta = -\big(\sqrt{2}+\cos\theta+\sin\theta\big).
\end{align}
De certa forma, a violação da desigualdade CHSH-Bell está relacionada com a incompatibilidade dos observáveis locais.

## Quantificação de não localidade
Como podemos quantificar a não localidade de um estado bipartido $\rho_{AB}$? Até onde sei, tem-se focado na não localidade o estado, obtida em geral maximizando uma testemunha de não localidade sobre os observáveis locais.

Por exemplo, seguindo (https://arxiv.org/abs/1510.08030v3), escrevemos um operador densidade de 2 qubits em termos das bases locais de Pauli:
$$4\rho_{AB} = I\otimes I +\vec{x}\cdot\vec{\sigma}\otimes I + I\otimes\vec{y}\cdot\vec{\sigma} + \sum_{j,k=1}^{3}c_{jk}\sigma_{j}\otimes\sigma_{k}.$$
A não localidade deste estado é dada por
$$N(\rho_{AB}) = \max\left(0,\frac{\sqrt{c^{2}-c_{\min}^{2}}-1}{\sqrt{2}-1}\right),$$
em $c_{\min}$ é a menor das componentes do vetor de correlação $\vec{c}=(c_{1},c_{2},c_{3})$, $c=||\vec{c}||$. Usamos como componentes de $\vec{c}$ os autovalores da matriz de correlação $C=(c_{jk})$.

```python
def chsh(rho):
    cm = zeros(3,3); cm = corr_mat(2, 2, rho)
    evals = zeros(3,1); W = zeros(3,1); W = eVals(3, cm)
    return max(0,(sqrt(W[0]**2+W[1]**2+W[2]**2-min(W[0],W[1],W[2])**2)-1)/(sqrt(2)-1))
```
